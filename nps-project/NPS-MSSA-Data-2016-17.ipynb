```python
def get_loc_week():
    dfs = list()
    for files in WEEK1_7_FILES:
        file_split = files.split(" ")
        week = int(file_split[2])
        location = file_split[5].split(".")[0]
        df = pd.read_csv(files)
        df['week'] = week
        df['location'] = location
        dfs.append(df)
        
    return dfs

array_of_dfs = get_loc_week()
```
